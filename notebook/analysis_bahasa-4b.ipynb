```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data ANalysis

```python
df['is_correct'] = (df.preds == df.kunci)
```

```python
mark_lvl = df.groupby('level')['is_correct'].mean()
```

```python
levels = [
    'Kelas I SD',
    'Kelas II SD',
    'Kelas III SD',
    'Kelas IV SD',
    'V SD',
    'VI SD',
    'VII SMP',
    'VIII SMP ',
    'IX SMP',
    'Kelas X SMA',
    'Kelas XI SMA',
    'Kelas XII SMA',
    'Seleksi PTN',
    'SD',
    'SD-SMP',
    'SMA',
    'SD-SMP-SMA',
    'SMP',
]
mark_lvl = mark_lvl[levels]

df_mark_lvl = pd.DataFrame()
df_mark_lvl['mark'] = mark_lvl
df_mark_lvl.reset_index(inplace=True)

df_mark_lvl.loc[len(df_mark_lvl)] = ('[AVERAGE]', df['is_correct'].mean())

df_mark_lvl
```

```python
sns.barplot(data=df_mark_lvl, x='mark', y='level')
plt.xlim(0,1)
plt.title('Rerata Nilai per Level')
```

```python
mark_sub = df.groupby('subject')['is_correct'].mean()
df_mark_sub = pd.DataFrame()

df_mark_sub['mark'] = mark_sub
df_mark_sub.reset_index(inplace=True)
df_mark_sub.loc[len(df_mark_sub)] = ('[AVERAGE]', df['is_correct'].mean())
df_mark_sub.sort_values('mark', inplace=True, ascending=False)
df_mark_sub
```

```python
sns.barplot(data=df_mark_sub, x='mark', y='subject')
plt.xlim(0,1)
plt.title('ArithmeticError')
```

# perbadingan antara model sekarang dengan model lain

```python
subject_map = {
    'STEM': {'Biologi', 'Fisika', 'Kimia', 'IPA'},
    'SOCIAL SCIENCE': {'Ekonomi', 'Geografi', 'IPS', 'Sejarah', 'Sosiologi'},
    'HUMANITIES': {'Agama Hindu', 'Agama Islam', 'Agama Kristen', 'Kesenian', 'PPKN', 'Penjaskes'},
    'INDONESIAN LANG.': {'Bahasa Indonesia'},
    'LOCAL L. CULTURE': {'Bahasa Bali', 'Bahasa Banjar', 'Bahasa Dayak Ngaju', 'Bahasa Jawa', 'Bahasa Lampung', 'Bahasa Madura', 'Bahasa Makassar', 'Bahasa Sunda', 'Budaya Alam Minangkabau'}
}
```

```python
# Sanity check
subjects = list(mark_sub.index)
inverse_map = {}

for subject in subjects:
    containing = False
    for category, sub_set in subject_map.items():
        if subject in sub_set:
            inverse_map[subject] = category
            containing = True
            break
    if not containing:
        inverse_map[subject] = None

inverse_map
```

```python
df['subject_category'] = df['subject'].apply(lambda s: inverse_map[s])
df.head()
```

```python
# Compute score per category
mark_cat = df.groupby('subject_category')['is_correct'].mean() * 100
df_mark_cat = pd.DataFrame()

df_mark_cat['mark'] = mark_cat
df_mark_cat.reset_index(inplace=True)
df_mark_cat.loc[len(df_mark_cat)] = ('AVERAGE', df['is_correct'].mean()*100)
df_mark_cat.sort_values('mark', inplace=True, ascending=False)
df_mark_cat.set_index('subject_category', inplace=True)
df_mark_cat
```

```python
('[Bahasa (4B)]()', *tuple(df_mark_cat.loc[['STEM','SOCIAL SCIENCE','HUMANITIES','INDONESIAN LANG.','LOCAL L. CULTURE','AVERAGE']].mark.values))
```

```python
models_bench = pd.read_csv('../data/IndoMMLU_Benchmark.csv')
models_bench.loc[len(models_bench)] = (
    '[Bahasa (4B)]()', 
    *tuple(df_mark_cat.loc[['STEM','SOCIAL SCIENCE','HUMANITIES','INDONESIAN LANG.','LOCAL L. CULTURE','AVERAGE']].mark.values)
)
models_bench
```

```python
models_bench.sort_values('Average', ascending=False)
```

```python
# Split model name and url
import re

pattern = r'\[(.*)\]\((.*)\)'
s = '[GPT-3.5 (175B)](https://chat.openai.com/)'

splitted = models_bench['Model (#param)'].apply(lambda s: re.findall(pattern, s)[0])

model_name = splitted.apply(lambda t: t[0])
model_url = splitted.apply(lambda t: t[1])
```

```python
models_bench['Model'] = model_name
models_bench['Repo'] = model_url
result = models_bench.drop(['Model (#param)'], axis=1)
result = result[['Model', 'Repo', *list(result.columns)[:-2]]]
result["Model"] = result['Model'].apply(lambda s: re.sub(r"<sub>(.*?)</sub>", r"_\1", s)).values

result
```

```python
result_sorted = result.sort_values('Average', ascending=False)
result_sorted
```

```python
# Save to csv
result_sorted.to_csv('bahasa_4b_indo_mmlu_benchmarking.csv', index=False)
```
